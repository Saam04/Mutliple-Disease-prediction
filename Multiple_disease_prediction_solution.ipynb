```python
y = disease['prognosis']
```

```python
x = disease[['itching','skin_rash','continuous_sneezing','shivering','chills','joint_pain','stomach_pain','acidity','ulcers_on_tongue','blackheads','scurring','skin_peeling','silver_like_dusting','small_dents_in_nails','inflammatory_nails','blister','red_sore_around_nose','yellow_crust_ooze','nodal_skin_eruptions']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
lr.predict_proba(x_test)
```
